```python
def add_lag_features(
    df,
    group_cols=["Disease_group", "Reporting Area norm"],
    value_col="Current week"
):
    df = df.sort_values(group_cols + ["Current MMWR Year", "MMWR WEEK"]).copy()

    g = df.groupby(group_cols)[value_col]

    # Lags: depend only on the past
    for lag in [1, 2, 3, 4, 8, 12, 16]:
        df[f"lag_{lag}"] = g.shift(lag)

    # Rolling stats over HISTORY ONLY (up to t-1)
    past = g.shift(1)

    # (Your approach works; keeping it consistent with what you wrote)
    key = df[group_cols].apply(tuple, axis=1)

    df["roll_mean_4"] = past.groupby(key).transform(lambda s: s.rolling(4, min_periods=4).mean())
    df["roll_mean_8"] = past.groupby(key).transform(lambda s: s.rolling(8, min_periods=8).mean())
    df["roll_std_4"]  = past.groupby(key).transform(lambda s: s.rolling(4, min_periods=4).std())
    df["roll_std_8"]  = past.groupby(key).transform(lambda s: s.rolling(8, min_periods=8).std())

    # Momentum / growth from past only (no current value used)
    # growth_1 as of last week: x_{t-1} - x_{t-2}
    df["growth_1"] = df["lag_1"] - df["lag_2"]

    # growth_4 as of last week: x_{t-1} - x_{t-5}
    lag_5 = g.shift(5)
    df["growth_4"] = df["lag_1"] - lag_5

    # Percent change versions (also past-only)
    df["pct_change_1"] = df["growth_1"] / (df["lag_2"].abs() + 1e-6)
    df["pct_change_4"] = df["growth_4"] / (lag_5.abs() + 1e-6)

    return df

def add_time_features(df):
    df = df.copy()

    # ensure ints
    df["year"] = df["Current MMWR Year"].astype(int)
    df["week"] = df["MMWR WEEK"].astype(int)

    # Build an ISO-week date (Monday of that week)
    iso_str = df["year"].astype(str) + "-W" + df["week"].astype(str).str.zfill(2) + "-1"
    df["week_start_date"] = pd.to_datetime(iso_str, format="%G-W%V-%u", errors="coerce")

    # calendar features (numeric)
    df["month"] = df["week_start_date"].dt.month.astype("Int64")
    df["quarter"] = df["week_start_date"].dt.quarter.astype("Int64")
    df["day_of_year"] = df["week_start_date"].dt.dayofyear.astype("Int64")

    # seasons from week number
    df["season"] = pd.cut(
        df["week"],
        bins=[0, 13, 26, 39, 53],
        labels=["winter", "spring", "summer", "fall"],
        include_lowest=True
    )

    # Create dummies, then force consistent columns + int dtype
    season_dummies = pd.get_dummies(df["season"], prefix="season")

    # Ensure all four season columns exist (important for train/test consistency)
    for col in ["season_winter", "season_spring", "season_summer", "season_fall"]:
        if col not in season_dummies.columns:
            season_dummies[col] = 0

    season_dummies = season_dummies[["season_winter", "season_spring", "season_summer", "season_fall"]].astype(int)

    # Attach and drop the original season column
    df = pd.concat([df.drop(columns=["season"]), season_dummies], axis=1)

    return df
```

```python
us_neighbors = {
    "Alabama": ["Florida", "Georgia", "Mississippi", "Tennessee"],
    "Arizona": ["California", "Nevada", "Utah", "Colorado", "New Mexico"],
    "Arkansas": ["Missouri", "Tennessee", "Mississippi", "Louisiana", "Texas", "Oklahoma"],
    "California": ["Oregon", "Nevada", "Arizona"],
    "Colorado": ["Wyoming", "Nebraska", "Kansas", "Oklahoma", "New Mexico", "Arizona", "Utah"],
    "Connecticut": ["New York", "Massachusetts", "Rhode Island"],
    "Delaware": ["Maryland", "New Jersey", "Pennsylvania"],
    "Florida": ["Alabama", "Georgia"],
    "Georgia": ["Florida", "Alabama", "Tennessee", "North Carolina", "South Carolina"],
    "Idaho": ["Washington", "Oregon", "Nevada", "Utah", "Wyoming", "Montana"],
    "Illinois": ["Indiana", "Kentucky", "Missouri", "Iowa", "Wisconsin"],
    "Indiana": ["Illinois", "Michigan", "Ohio", "Kentucky"],
    "Iowa": ["Minnesota", "South Dakota", "Nebraska", "Missouri", "Illinois", "Wisconsin"],
    "Kansas": ["Nebraska", "Missouri", "Oklahoma", "Colorado"],
    "Kentucky": ["Indiana", "Ohio", "West Virginia", "Virginia", "Tennessee", "Missouri", "Illinois"],
    "Louisiana": ["Texas", "Arkansas", "Mississippi"],
    "Maine": ["New Hampshire"],
    "Maryland": ["Virginia", "West Virginia", "Pennsylvania", "Delaware"],
    "Massachusetts": ["Connecticut", "Rhode Island", "New York", "New Hampshire", "Vermont"],
    "Michigan": ["Ohio", "Indiana", "Wisconsin"],
    "Minnesota": ["North Dakota", "South Dakota", "Iowa", "Wisconsin"],
    "Mississippi": ["Louisiana", "Arkansas", "Tennessee", "Alabama"],
    "Missouri": ["Kansas", "Oklahoma", "Arkansas", "Tennessee", "Kentucky", "Illinois", "Iowa", "Nebraska"],
    "Montana": ["Idaho", "Wyoming", "South Dakota", "North Dakota"],
    "Nebraska": ["South Dakota", "Iowa", "Missouri", "Kansas", "Colorado", "Wyoming"],
    "Nevada": ["California", "Oregon", "Idaho", "Utah", "Arizona"],
    "New Hampshire": ["Vermont", "Maine", "Massachusetts"],
    "New Jersey": ["New York", "Delaware", "Pennsylvania"],
    "New Mexico": ["Arizona", "Utah", "Colorado", "Oklahoma", "Texas"],
    "New York": ["Pennsylvania", "New Jersey", "Connecticut", "Massachusetts", "Vermont"],
    "North Carolina": ["South Carolina", "Georgia", "Tennessee", "Virginia"],
    "North Dakota": ["Montana", "South Dakota", "Minnesota"],
    "Ohio": ["Michigan", "Indiana", "Kentucky", "West Virginia", "Pennsylvania"],
    "Oklahoma": ["Texas", "Kansas", "Missouri", "Arkansas", "New Mexico", "Colorado"],
    "Oregon": ["California", "Nevada", "Idaho", "Washington"],
    "Pennsylvania": ["New York", "New Jersey", "Delaware", "Maryland", "West Virginia", "Ohio"],
    "Rhode Island": ["Connecticut", "Massachusetts"],
    "South Carolina": ["Georgia", "North Carolina"],
    "South Dakota": ["North Dakota", "Minnesota", "Iowa", "Nebraska", "Wyoming", "Montana"],
    "Tennessee": ["Kentucky", "Virginia", "North Carolina", "Georgia", "Alabama", "Mississippi", "Arkansas", "Missouri"],
    "Texas": ["New Mexico", "Oklahoma", "Arkansas", "Louisiana"],
    "Utah": ["Idaho", "Wyoming", "Colorado", "New Mexico", "Arizona", "Nevada"],
    "Vermont": ["New Hampshire", "Massachusetts", "New York"],
    "Virginia": ["North Carolina", "Tennessee", "Kentucky", "West Virginia", "Maryland"],
    "Washington": ["Idaho", "Oregon"],
    "West Virginia": ["Ohio", "Pennsylvania", "Maryland", "Virginia", "Kentucky"],
    "Wisconsin": ["Minnesota", "Iowa", "Illinois", "Michigan"],
    "Wyoming": ["Montana", "South Dakota", "Nebraska", "Colorado", "Utah", "Idaho"]
}


def add_neighbor_features(df, neighbors_dict, case_col="Current week"):
    # Make a copy and ensure we have normalized state names
    df = df.copy()

    # Sort for consistency
    df = df.sort_values(["Disease_group", "Reporting Area norm", "year", "week"])

    # Build an index for fast lookup: (state, year, week) -> case count(s)
    df_key = df.set_index(["Reporting Area norm", "year", "week"])

    neighbor_sum = []
    neighbor_mean = []
    neighbor_max = []
    neighbor_count = []

    for idx, row in df.iterrows():
        state = row["Reporting Area norm"]
        year = row["year"]
        week = row["week"]

        neighbors = neighbors_dict.get(state, [])

        neigh_values = []

        for n in neighbors:
            try:
                vals = df_key.loc[(n, year, week), case_col]
                # If multiple rows match, vals is a Series -> aggregate
                if isinstance(vals, pd.Series):
                    val = float(vals.sum())
                else:
                    val = float(vals)
                neigh_values.append(val)
            except KeyError:
                # No data for that neighbor in that week/year
                continue

        if not neigh_values:
            # If no neighbor data, treat as zeros
            neigh_values = [0.0]

        arr = np.array(neigh_values, dtype=float)
        neighbor_sum.append(arr.sum())
        neighbor_mean.append(arr.mean())
        neighbor_max.append(arr.max())
        neighbor_count.append((arr > 0).sum())

    df["neighbor_cases_sum"] = neighbor_sum
    df["neighbor_cases_mean"] = neighbor_mean
    df["neighbor_cases_max"] = neighbor_max
    df["neighbor_states_reporting"] = neighbor_count

    return df
```
